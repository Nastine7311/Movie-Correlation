```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adust the configuration of the plot we will create


# Read the Data

df = pd.read_csv('movies 1.csv')
```

```python
#Let's view the Data

df.head()
```

```python
# Check for missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#check data types

df.dtypes
```

```python
#check data shape

df.shape
```

```python
missing_values_budget = df['budget'].isnull().sum()
print("Number of missing values in 'budget' column:", missing_values_budget)
```

```python
# View rows with missing values in both the 'budget' and 'gross' columns
missing_values_df = df[df['budget'].isnull() | df['gross'].isnull()]

# Print the rows with missing values in both columns
print("Rows with missing values in 'budget' or 'gross' column:")
print(missing_values_df)
```

```python
# Change data type of multiple columns and replace missing values with zero
columns_to_convert = ['budget', 'gross', 'votes', 'runtime']
for col in columns_to_convert:
    df[col] = df[col].fillna(0).astype('int64')
```

```python
df.head()
```

```python
# Correct year column

# Fill missing values in 'released' column with a default value
df['released'].fillna('1900-01-01', inplace=True)
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# To view full data

pd.set_option('display.max_rows', None)
```

```python
# Check for duplicates
duplicates = df[df.duplicated()]
df.head()
```
